```python
import pandas as pd
import plotly.express as px

# For Google Colab: upload file
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Read the file
df = pd.read_csv(filename)
df['BRICS Country'] = df['BRICS Country'].astype(str).str.strip()
df['Gulf Country'] = df['Gulf Country'].astype(str).str.strip()
df['Sector'] = df['Sector'].astype(str).str.strip().str.lower()

# Group sectors
def map_sector(s):
    if 'oil' in s:
        return 'Oil'
    if 'gas' in s or 'lng' in s:
        return 'Gas'
    if 'coal' in s:
        return 'Coal'
    if ('renew' in s or 'solar' in s or 'wind' in s or
        'hydro' in s or 'geoterm' in s or 'bio' in s):
        return 'Renewables'
    if 'nuclear' in s:
        return 'Nuclear'
    if ('infrastr' in s or 'infra' in s or 'grid' in s or 'storage' in s or 'r&d' in s or
        'refinery' in s or 'service' in s or 'pipeline' in s or 'power' in s):
        return 'Infrastructure/Other'
    return 'Infrastructure/Other'

df['Sector_grouped'] = df['Sector'].apply(map_sector)

# Group and count
df_grouped = (
    df.groupby(['BRICS Country', 'Gulf Country', 'Sector_grouped'])
      .size()
      .reset_index(name='Project Count')
)

# SUNBURST
fig1 = px.sunburst(
    df_grouped,
    path=['BRICS Country', 'Gulf Country', 'Sector_grouped'],
    values='Project Count',
    color='BRICS Country',
    title="Regional preferences: BRICS → Gulf Country → Sector (Sunburst)"
)
fig1.update_layout(margin=dict(t=60, l=0, r=0, b=0))
fig1.show()

# TREEMAP
fig2 = px.treemap(
    df_grouped,
    path=['BRICS Country', 'Gulf Country', 'Sector_grouped'],
    values='Project Count',
    color='BRICS Country',
    title="Regional preferences: BRICS → Gulf Country → Sector (Treemap)"
)
fig2.update_layout(margin=dict(t=60, l=0, r=0, b=0))
fig2.show()
```
